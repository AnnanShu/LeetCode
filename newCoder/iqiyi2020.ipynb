有一个x*y*z的立方体，要在这个立方体上砍k刀，每一刀可以看作是用一个平行于立方体某一面的平面切割立方体，且必须在坐标为整数的位置切割，如在x=0.5处用平面切割是非法的。

问在切割k刀之后，最多可以把立方体切割成多少块。

```python
import sys
x, y, z, k = list(map(int, sys.stdin.readline().split()))
x, y, z = sorted([x, y, z])
x_i, y_i, z_i = 1, 1, 1
x_fill, y_fill, z_fill = [False] * 3
if k >= (x + y + z) - 3: 
    print(x*y*z) 
else:
    for i in range(k):
        if not x_fill:
            if x_i == y_i == z_i:
                x_i += 1
                if x_i == x:
                    x_fill = True 
            elif y_i == z_i:
                y_i += 1
            else:
                z_i += 1
        elif not y_fill:
            if y_i == z_i:
                y_i += 1
                if y_i == y:
                    y_fill = True
            else:
                z_i += 1
        else:
            z_i += 1
            
    print(x_i * y_i * z_i)
```

```python
def build(p):
    m = len(p)
    nxt = [0, 0]
    j = 0
    for i in range(1, m):
        while j > 0 and p[i] != p[j]: 
            j = nxt[j]
        if p[i] == p[j]: 
            j += 1
        nxt.append(j)
    return nxt
import sys 
s = sys.stdin.readline().strip() 
t = sys.stdin.readline().strip()
n, m = len(s), len(t)
nxt = build(t)
res = [0] * n 
j = 0
for i in range(n):
    while j > 0 and s[i] != t[j]:
        j = nxt[j] 
    if s[i] == t[j]:
        j += 1
    if j == m:
        if i == 0:
            res[i] = 1
        else:
            res[i] = res[i-1] + 1
        j = nxt[j]
    else:
        if i == 0:
            res[i] = 0
        else:
            res[i] = res[i-1]

print(" ".join([str(each) for each in res]))
```
